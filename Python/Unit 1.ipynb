```python
# Series : column in table , 1D array holding data of any type

import pandas as pd

a = [1,7,2]
myvar = pd.Series(a)
print(myvar)
# 0    1
# 1    7
# 2    2
# dtype: int64
print(myvar[2]) #2
# print(myvar[7]) KeyError: 7

myvar2 = pd.Series(a,index=['x','y','z'])
print(myvar2)
# x    1
# y    7
# z    2
# dtype: int64
print(myvar2['y']) #7
# print(myvar2['a']) KeyError: 'a'

b = [1.4,5,6,7]
var = pd.Series(b)
print(var)
# 0    1.4
# 1    5.0
# 2    6.0
# 3    7.0
# dtype: float64


c = [1.4,"5",6,7]
var1 = pd.Series(c)
print(var1)
# 0    1.4
# 1      5
# 2      6
# 3      7
# dtype: object
```

```python
import pandas as pd
student_name=["Ram","Shyam","Radha","Sita","Geeta"]
marks=[50,60,70,80,90]
varR = pd.Series(marks,index=student_name,name="Results")
print(varR)
# Ram      50
# Shyam    60
# Radha    70
# Sita     80
# Geeta    90
# Name: Results, dtype: int64
```

```python
import pandas as pd
calories = {'day1':400,'day2':350,'day3':430}
varC = pd.Series(calories)
print(varC)
# day1    400
# day2    350
# day3    430
# dtype: int64
varC = pd.Series(calories,index=['day1','day2'])
print(varC)
# day1    400
# day2    350
# dtype: int64
varC = pd.Series(calories,index=['d1','d2'])
print(varC)
# d1   NaN
# d2   NaN
# dtype: float64
```

```python
import pandas as pd
marks = {'Ram':80,'Shyam':70,'Seeta':78,'Geeta':90}
res = pd.Series(marks)

print(res)
# Ram      80
# Shyam    70
# Seeta    78
# Geeta    90
# dtype: int64

print(res.size) # number of entries --> Here : 4

print(res.dtype) # int64

print(res.name) # None

print(res.index) # Index(['Ram', 'Shyam', 'Seeta', 'Geeta'], dtype='object')

print(res.values) # [80 70 78 90] --> ARRAY (no commas)

print(res.is_unique) # True --> True or False according to the data if unique values : True
```

```python
import pandas as pd
subs = pd.read_csv("subs.csv") # DataFrame

print(subs)
#      Subscribers gained
# 0                    48
# 1                    57
# 2                    40
# 3                    43
# 4                    44
# ..                  ...
# 360                 231
# 361                 226
# 362                 155
# 363                 144
# 364                 172

# [365 rows x 1 columns]

print(type(subs))
# <class 'pandas.core.frame.DataFrame'>

print(subs.info())
# memory usage: 3.0 KB
# None
```

```python
import pandas as pd
subs = pd.read_csv("subs.csv",squeeze=True) # Sqeeze works on Single column ONLY --> converts to SERIES

print(subs)
# 0       48
# 1       57
# 2       40
# 3       43
# 4       44
#       ... 
# 360    231
# 361    226
# 362    155
# 363    144
# 364    172
# Name: Subscribers gained, Length: 365, dtype: int64

print(type(subs)) #<class 'pandas.core.series.Series'>

subs.describe()
# count    365.000000
# mean     135.643836
# std       62.675023
# min       33.000000
# 25%       88.000000
# 50%      123.000000
# 75%      177.000000
# max      396.000000
# Name: Subscribers gained, dtype: float64

# 25%, 50% and 75% tells the values at 25%, 50% and 75% th index after arranging it in ascending order --> used in IQR
# IQR : interquartile range

print(subs.min()) # 33
print(subs.max()) # 396
print(subs.median()) # 123.0
print(subs.mean()) # 135.64383561643837
print(subs.sum()) # 49510

print(subs[subs>200])
# 165    225
# 166    249
# 167    265
# 168    306
# 169    261
# 170    222
# 225    224
# 226    254
print(subs[subs>200].size) # 59
print(subs[subs>200].count()) # 59
```
