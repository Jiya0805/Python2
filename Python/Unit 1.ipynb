```python
# Series : column in table , 1D array holding data of any type

import pandas as pd

a = [1,7,2]
myvar = pd.Series(a)
print(myvar)
# 0    1
# 1    7
# 2    2
# dtype: int64
print(myvar[2]) #2
# print(myvar[7]) KeyError: 7

myvar2 = pd.Series(a,index=['x','y','z'])
print(myvar2)
# x    1
# y    7
# z    2
# dtype: int64
print(myvar2['y']) #7
# print(myvar2['a']) KeyError: 'a'

b = [1.4,5,6,7]
var = pd.Series(b)
print(var)
# 0    1.4
# 1    5.0
# 2    6.0
# 3    7.0
# dtype: float64


c = [1.4,"5",6,7]
var1 = pd.Series(c)
print(var1)
# 0    1.4
# 1      5
# 2      6
# 3      7
# dtype: object
```
