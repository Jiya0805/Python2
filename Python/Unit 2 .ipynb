```python
import numpy as np
import pandas as pd
time = np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color="green")
plt.plot(time,expense,color="red")
plt.fill_between(time,income,expense,where=(income>expense),color="green",alpha=0.25,label='postive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color="red",alpha=0.25,label='negative',interpolate=True)
plt.xlabel('time')
plt.ylabel('income/expense')
plt.legend()
plt.show()
```

```python
from wordcloud import WordCloud,STOPWORDS

alice_novel =open('alice_in_wonderland.txt')
alice_novel=alice_novel.read()

stopwords= set(STOPWORDS)
# print(stopwords)
print(len(stopwords))

alice_wc = WordCloud(background_color='white',max_words=2000,stopwords=stopwords)
alice_wc.generate(alice_novel)
plt.axis('off')
plt.imshow(alice_wc)
plt.show()
```
