```python
doo = pd.read_csv('auto-mpg.csv')

def find_outliers(ds,col):
    q1 = ds[col].quantile(0.25)
    q3 = ds[col].quantile(0.75)
    IQR = q3-q1
    low= q1-1.5*IQR
    high= q3+1.5*IQR  
    print(low)
    print(high)
    
    ds = ds.loc[(ds[col]<low) | (ds[col]>high) ]
    return ds

find_outliers(doo,"mpg")
```

```python
def remove_outliers(ds,col):
    q1 = ds[col].quantile(0.25)
    q3 = ds[col].quantile(0.75)
    IQR = q3-q1
    low= q1-1.5*IQR
    high= q3+1.5*IQR  
    print(low)
    print(high)
    
    ds = ds.loc[(ds[col]>=low) & (ds[col]<=high) ]
    return ds

remove_outliers(doo,"acceleration")
```

```python
time = np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color="green")
plt.plot(time,expense,color="red")
plt.fill_between(time,income,expense,where=(income>expense),color="green",alpha=0.25,label='postive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color="red",alpha=0.25,label='negative',interpolate=True)
plt.xlabel('time')
plt.ylabel('income/expense')
plt.legend()
plt.show()
```
